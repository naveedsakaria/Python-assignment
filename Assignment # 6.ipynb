# Q # 1. Define Object Oriented Programming Language?

# Answer : Object-oriented programming (OOP) refers to a type of computer programming (software design) in which programmers define the data type of a data structure, and also the types of operations (functions) that can be applied to the data structure.

# In this way, the data structure becomes an object that includes both data and functions. In addition, programmers can create relationships between one object and another. For example, objects can inherit characteristics from other objects.

# Q#3. Differentiate between function and method?


#       F U N C T I O N 
## Function is block of code that is also called by its name. (independent)
## The function can have different parameters or may not have any at all. If any data (parameters) are passed, they are passed explicitly.
## It may or may not return any data.
## Function does not deal with Class and its instance concept.


#        M E T H O D
## Method is called by its name, but it is associated to an object (dependent).
## A method is implicitly passed the object on which it is invoked.
## It may or may not return any data.
## A method can operate on the data (instance variables) that is contained by the  corresponding class

```python
# Q#5. Write a code in python in which create a class named it Car which 
# have 5 attributes such like (model, color and name etc.) and 3
# methods. And create 5 object instance from that class.

# Answer

class car():
    # Attributes
    def __init__(self,model,color,name,maker,vehicleno):
        self.model = model
        self.color = color
        self.name = name
        self.maker = maker
        self.vehicleno = vehicleno
    
    # Methods
    def start(self):
        print("Engine Starts")
        
    def stop(self):
        print("Engine Off")
        
    def gear(self):
        print("Car is Moving Fast")
    
# Instance
car1 = car(2019,'red','BMW','Mercides','V1201')
print("xxxxxxx Car 1 xxxxxxx")
print(car1.maker)
print(car1.name)
print(car1.color)
print(car1.model)
print(car1.vehicleno)
car1.start()
car1.stop()
car1.gear()
print("xxxxxxxxxxxxxxxxxxxxx\n")


car2 = car(2018,'blue','corolla','toyota','v1202')
print("xxxxxxx Car 2 xxxxxxx")
print(car2.maker)
print(car2.name)
print(car2.color)
print(car2.model)
print(car2.vehicleno)
car2.start()
car2.stop()
car2.gear()
print("xxxxxxxxxxxxxxxxxxxxx\n")

car3 = car(2017,'white','civic','honda','v1203')
print("xxxxxxx Car 3 xxxxxxx")
print(car3.maker)
print(car3.name)
print(car3.color)
print(car3.model)
print(car3.vehicleno)
car3.start()
car3.stop()
car3.gear()
print("xxxxxxxxxxxxxxxxxxxxx\n")

car4 = car(2016,'black','premieo','honda','v1204')
print("xxxxxxx Car 4 xxxxxxx")
print(car4.maker)
print(car4.name)
print(car4.color)
print(car4.model)
print(car4.vehicleno)
car4.start()
car4.stop()
car4.gear()
print("xxxxxxxxxxxxxxxxxxxxx\n")

car5 = car(2015,'silver','cultus','suzuki','v1205')
print("xxxxxxx Car 5 xxxxxxx")
print(car5.maker)
print(car5.name)
print(car5.color)
print(car5.model)
print(car5.vehicleno)
car5.start()
car5.stop()
car5.gear()
print("xxxxxxxxxxxxxxxxxxxxx\n")
```
